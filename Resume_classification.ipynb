# Preprocessing

```python
rdf = rdf.drop(['ID', 'Resume_html'], axis=1)
```

```python
rdf.isna().sum()
```

```python
na_rows = rdf.isna().any(axis=1)

# Display the indices of rows with missing values
missing_value_indices = rdf[na_rows].index.tolist()
print(missing_value_indices)
```

```python
for i in missing_value_indices:
    print(rdf.loc[i])
```

We will remove the rows with NaN in Resume_str and Category

```python
na_rows = rdf[['Resume_str', 'Category']].isna().any(axis=1)
```

```python
rdf = rdf.drop(rdf[na_rows].index)
```

```python
rdf.isna().sum()
```

```python
rdf.shape
```

We will make seperate columns for heading, Highlights, Experience, Education and skills that are extracted from the Resume_str column and finally dropping the Resume_str column

```python
rdf = rdf.rename(columns={"Resume_str": "Resume"}, errors="raise")
```

```python
rdf.info()
```

```python
# Count the occurrences of each category
category_counts = rdf['Category'].value_counts()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")

# Set title and labels
plt.title('Counts of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate labels for better readability if needed

# Display the plot
plt.tight_layout()
plt.show()
```
